```python
import numpy as np
import pandas as pd
```

```python
data={
    "Name": ["John","Sarah","Mike","John","Indai"],
    "Age": [25,None,-10,25,40],
    "Gender": ["M","Female","male","M","F"],
    "Salary": [50000,52000,None,50000,70000]
}


df=pd.DataFrame(data)
print(df)
```

```python
print(df.isnull())
```

```python
print(df.isnull().sum())
```

```python
# checking for duplicate rows
print(df.duplicated())

# count the duplicated rows
print(df.duplicated().sum())

print(df[df.duplicated()])
```

```python
print(df.dtypes)
```

```python
new = df.drop_duplicates()
print(new)
```

```python
#replace NaN values with mean/median/mode
new = df.copy()
new['Age'] = new['Age'].fillna(new['Age'].mean())
new['Salary'] = new['Salary'].fillna(new['Salary'].mean())
new
```

```python
# Safe copy of df
new = df.copy()

# Replace NaN in numeric columns
new['Age'] = new['Age'].fillna(new['Age'].mean())
new['Salary'] = new['Salary'].fillna(new['Salary'].mean())

# Standardize gender values
new['Gender'] = new['Gender'].replace({"M": "Male", "F": "Female", "m": "Male"})
new
```

```python
#applying the lambda function
new['Age'] = new['Age'].apply(lambda x: max(0, x))
new
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()


columns_to_scale = ['Age','Salary']
new[columns_to_scale] = scaler.fit_transform(new[columns_to_scale])

new
```
